```python
from keras.datasets import imdb
from keras import Sequential
from keras.preprocessing.text import Tokenizer
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(X_train, y_train), (X_test, y_test) = imdb.load_data()
```

```python
X_train.shape
```

```python
# X_train[0]
```

```python
from keras.utils import pad_sequences
X_train = pad_sequences(X_train, padding='post', maxlen=50)
X_test = pad_sequences(X_test, padding = 'post', maxlen=50)
# data is shorten
```

```python
X_train.shape
```

```python
X_train[0]
```

```python
model = Sequential()

model.add(SimpleRNN(32, input_shape=(50,1), return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs = 5, validation_data=(X_test, y_test))
```

```python
# Now lets use Embedding on above code .
# data should be integer encoding
```

```python
model = Sequential()

model.add(Embedding(88586,2)) # vocab size = 10k, required dense vector = 2D, input_size = 50
model.add(SimpleRNN(32, return_sequences=False)) # 32 nodes, and o/p will be come out
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
import numpy as np

# Assuming X_train and X_test are your input data
print(f"Max index in X_train: {np.max(X_train)}")
print(f"Max index in X_test: {np.max(X_test)}")

# Ensure all indices are within the vocabulary size
X_train = np.clip(X_train, 0, 9999)
X_test = np.clip(X_test, 0, 9999)
```

```python
model.compile(optimizer = 'adam',loss='binary_crossentropy', metrics=['acc'])
history = model.fit(X_train, y_train, epochs = 5, validation_data=(X_test, y_test))
```
